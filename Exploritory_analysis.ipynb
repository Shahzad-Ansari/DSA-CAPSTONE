```python
import time
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapiExercises",timeout=200)

def city_state_country(row):
    global counter, start_time
    counter += 1
    elapsed_time = time.time() - start_time
    if counter % 10 == 0:
      print("Time elapsed for iteration {}: {:.2f} seconds".format(counter, elapsed_time))
      print("Time per iteration: {:.2f} seconds".format(elapsed_time / counter))
    coord = f"{row['lat']}, {row['lon']}"
    location = geolocator.reverse(coord, exactly_one=True)
    address = location.raw['address']
    city = address.get('city', '')
    state = address.get('state', '')
    county = address.get('county','')
    zip = address.get('postcode','')
    row['city'] = city
    row['state'] = state
    row['county'] = county
    row['zip'] = zip
    return row
```

```python

df = pd.read_excel('/content/SurveryData_with_locations.xlsx')
df = df.loc[df['state'] == 'Oklahoma']
locations = df[[ 'state','county']]
locations = locations.drop_duplicates().reset_index(drop=True)
counties = locations['county'].drop_duplicates().to_frame()
counties = counties.loc[counties['county'] != '']
counties['county_state'] = counties['county'] + ", Oklahoma"
```

```python
from geopy.geocoders import Nominatim

# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")
lat = []
long = []

for county in counties['county_state']:
  location = geolocator.geocode(county)
  lat.append(location.latitude)
  long.append(location.longitude)

counties = counties.drop(['county_state'], axis=1)
```

```python

counties['lat'] = lat
counties['long'] = long
counties.sort_values('county').reset_index(drop=True)
display(counties)
counties.to_excel('counties.xlsx')
```

```python

from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily

# Set time period
start = datetime(2018, 1, 1)
end = datetime(2018, 12, 31)
weatherData = []

for index, row in counties.iterrows():
  location = Point(row['lat'],row['long'])
  data = Daily(location, start, end)
  data = data.fetch()
  weatherData.append(data)
```

```python
dropList =[]
for idx, x in enumerate(weatherData):
  if x.empty:
    dropList.append(idx)
```

```python
weatherInformationByCounty = counties
weatherInformationByCounty['weatherData'] = weatherData
weatherInformationByCounty['county'] = locations['county']
weatherInformationByCounty['state'] = locations['state']
weatherInformationByCounty = weatherInformationByCounty.drop(dropList)
weatherInformationByCounty.sort_values('county').reset_index(drop=True)
```
